```python
#Usama Ansari (221206)
# ===============================
# STEP 1: Import Libraries
# ===============================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files

# ===============================
# STEP 2: Upload & Load Dataset
# ===============================
uploaded = files.upload()   # Upload Tweets.csv
data = pd.read_csv("Tweets.csv")

# ===============================
# STEP 3: Basic Information
# ===============================
print("Dataset Shape:", data.shape)
print("\nColumn Names:")
print(data.columns)

# ===============================
# STEP 4: Histogram (Distribution)
# ===============================
plt.figure(figsize=(6,4))
data.select_dtypes(include=['int64','float64']).iloc[:,0].hist()
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

# ===============================
# STEP 5: Line Chart
# ===============================
plt.figure(figsize=(6,4))
plt.plot(data.index, data.select_dtypes(include=['int64','float64']).iloc[:,0])
plt.title("Line Chart")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()

# ===============================
# STEP 6: Bar Chart
# ===============================
plt.figure(figsize=(6,4))
data.select_dtypes(include=['int64','float64']).iloc[:,0].head(10).plot(kind='bar')
plt.title("Bar Chart")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()

# ===============================
# STEP 7: Pie Chart
# ===============================
if len(data.select_dtypes(include=['object']).columns) > 0:
    data[data.select_dtypes(include=['object']).columns[0]].value_counts().head(5).plot(
        kind='pie', autopct='%1.1f%%'
    )
    plt.title("Pie Chart")
    plt.ylabel("")
    plt.show()

# ===============================
# STEP 8: Scatter Plot
# ===============================
num_cols = data.select_dtypes(include=['int64','float64']).columns
if len(num_cols) >= 2:
    plt.figure(figsize=(6,4))
    plt.scatter(data[num_cols[0]], data[num_cols[1]])
    plt.xlabel(num_cols[0])
    plt.ylabel(num_cols[1])
    plt.title("Scatter Plot")
    plt.show()
```
